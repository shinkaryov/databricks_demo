```python
schema_name = dbutils.widgets.get("schema_name")
source_table_name = dbutils.widgets.get("source_table_name")
source_table_name = dbutils.widgets.get("target_table_name")
```

```python
df = spark.read.table(f"{schema_name}.{source_table_name}")
```

```python
df = df.withColumnRenamed("FL_DATE", "date")
df = df.withColumnRenamed("AVG_TOTAL_DELAY", "avg_total_delay")
df = df.withColumnRenamed("MAX_ARR_DELAY", "max_arr_delay")
df = df.withColumnRenamed("FLIGHT_COUNT", "flight_count")
df.write.mode("overwrite").saveAsTable(f"{schema_name}.{target_table_name}")
```
